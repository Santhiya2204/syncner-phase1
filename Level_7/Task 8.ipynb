```python
import numpy as np
import matplotlib.pyplot as plt
n_trials = 10  
p_success = 0.5  
num_experiments = 1000  
data = np.random.binomial(n=n_trials, p=p_success, size=num_experiments)
mean_value = np.mean(data)
plt.hist(data, bins=range(12), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.title('Binomial Distribution Simulation')
plt.xticks(range(11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print(f"Mean of Simulated Distribution: {mean_value:.2f}")
print("Matches binomial expectation (n × p = 5)." if abs(mean_value - (n_trials * p_success)) < 0.5 else "Deviation observed from expected mean.")
```
