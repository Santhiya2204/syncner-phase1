```python
import networkx as nx
import matplotlib.pyplot as plt

dag = nx.DiGraph()

dag.add_edges_from([
    ("Task1", "Task2"),
    ("Task1", "Task3"),
    ("Task2", "Task4"),
    ("Task3", "Task4"),
    ("Task4", "Task5")
])

pos = nx.spring_layout(dag, seed=42)

plt.figure(figsize=(8, 6))
nx.draw(dag, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=12, font_weight="bold", arrows=True)
plt.title("DAG Visualization")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

dag = nx.DiGraph()

dag.add_edges_from([
    ("Ingest Data", "Clean Data"),
    ("Clean Data", "Transform Data"),
    ("Transform Data", "Model Training"),
    ("Model Training", "Model Evaluation"),
    ("Model Evaluation", "Deployment")
])

pos = nx.spring_layout(dag, seed=42)
plt.figure(figsize=(10, 6))
nx.draw(dag, pos, with_labels=True, node_size=3000, node_color="skyblue",
        font_size=10, font_weight="bold", arrows=True, edge_color='gray')
plt.title("DAG Visualization (Data Pipeline Example)")
plt.show()
```
